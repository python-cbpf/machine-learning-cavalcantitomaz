```
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop

# For a single-input model with 2 classes (binary classification):
model = Sequential()
model.add(Dense(1, activation='sigmoid', use_bias=True, input_dim=2))
model.add(Dense(3))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])


# Generate data
import numpy
import numpy as np
data = np.random.random((1500, 2))
labels = np.random.randint(2, size=(1500, 1))
        

# Train the model, iterating on the data in batches of 32 samples
model.fit(data, labels, epochs=100, batch_size=32)

# Plot training & validation accuracy values
history = model.fit(data, labels, validation_split=0.25, epochs=200, batch_size=32, verbose=1)

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()




```
